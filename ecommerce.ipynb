```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.month
df.groupby('Month')['TotalAmount'].sum()
```

```python
df['rating_segment'] = pd.cut(
    df['Rating'],
    bins=[0,2,3.5,5],
    labels=['Low','Medium','High']
)
df['rating_segment'].value_counts()
```

```python
corr = df[['Price','Quantity','TotalAmount','Rating','Reviews']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
category_revenue = df.groupby('Category')['TotalAmount'].sum()

plt.figure(figsize=(7,7))
plt.pie(
    category_revenue,
    labels=category_revenue.index,
    autopct='%1.1f%%',
   
)
plt.title("Revenue Contribution by Category")
plt.show()
```

```python
top_products = (
    df.groupby('Product')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
top_products.plot(kind='bar', figsize=(8,5))
plt.title("Top 10 Products by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Quantity'], bins=5, kde=True)
plt.title("Quantity Purchased Distribution")
plt.show()
```

```python
avg_order_value = df.groupby('Platform')['TotalAmount'].mean()

sns.barplot(
    x=avg_order_value.index,
    y=avg_order_value.values
)
plt.title("Average Order Value by Platform")
plt.ylabel("Avg Order Value")
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10)

top_cities.plot(kind='bar', figsize=(8,4))
plt.title("Top 10 Cities by Number of Orders")
plt.ylabel("Order Count")
plt.show()
```

```python
df['rating_segment'] = pd.cut(
    df['Rating'],
    bins=[0,2,3.5,5],
    labels=['Low','Medium','High']
)

sns.countplot(x='rating_segment', data=df)
plt.title("Customer Satisfaction Levels")
plt.show()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.month

monthly_revenue = df.groupby('Month')['TotalAmount'].sum()

plt.figure(figsize=(8,4))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
sns.boxplot(x='Platform', y='Rating', data=df)
plt.title("Rating Distribution by Platform")
plt.show()
```

```python
sns.boxplot(x=df['Reviews'])
plt.title("Reviews Distribution")
plt.show()
```

```python
df['price_segment'] = pd.cut(
    df['Price'],
    bins=[0,5000,10000,15000,20000],
    labels=['Low','Medium','High','Premium']
)

sns.countplot(x='price_segment', data=df)
plt.title("Product Price Segmentation")
plt.show()
```
